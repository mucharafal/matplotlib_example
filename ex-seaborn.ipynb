 ### Penguins
 Preloaded dataset from seaborn

 #### 1. Basic exploration

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins = penguins.dropna()
penguins.head()
```

```python
penguins.describe()
```

```python
g = sns.relplot(
    kind="scatter",
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="body_mass_g",
    style="species",
    palette="viridis_r")

g.fig.suptitle("Penguin weigth and bill dimmensions")
```

```python
g = sns.relplot(
    kind="scatter",
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="body_mass_g",
    col="species",
    style="sex",
    col_wrap=2,
    palette="viridis_r"
)
g.fig.suptitle("Penguin weigth and bill dimmensions +")
plt.subplots_adjust(top=0.9)
```

```python
g = sns.jointplot(data=penguins,
    x="flipper_length_mm",
    y="bill_length_mm",
    hue="species",
    palette="viridis_r")

g.fig.suptitle("Penguin flipper and bill sizes")
plt.subplots_adjust(top=0.9)
```
